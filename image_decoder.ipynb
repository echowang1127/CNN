```python
from scipy import misc
import tensorflow as tf

img = misc.imread('C:\\Users\\Public\\Pictures\\Sample Pictures\\Tulips.jpg')
print(img)
print (img.shape)   

img_tf = tf.Variable(img)
print (img_tf.get_shape().as_list() )
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
im = sess.run(img_tf)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
fig.add_subplot(1,2,1)
plt.imshow(im)
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.show()
```
